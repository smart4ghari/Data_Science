Example

* Draw a line in a diagram from position (0,0) to position (7,500):

```python
# Already we've imported matplotlib so just import numpy

import numpy as np

xpoints = np.array([0,7])
ypoints = np.array([0,500])

%matplotlib inline
plt.plot(xpoints,ypoints,marker='x')
plt.show()
```

### Plot function

* The plot function used to draw markers in a diagram
* By default the plot() draws a line from point to point
* This function takes input parameters as x and y axis 
* Parameter 1 is x axis
* Parameter 2 is y axis
* If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.


    - Here x axis is the Horizontal Axis
    - Here y axis is the Vertical Axis

```python
# Already we've imported numpy and matplotlib packages so don't import again and again just use the functions

x_pts = np.array([1,8]) # Here we've just used only x axis points
y_pts = np.array([3,10]) # Here we've just used only y axis points

plt.plot(x_pts,y_pts)
plt.show()
```

### Plotting Without Line
* To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
# Plotting the graph without line
plt.plot(x_pts,y_pts,'o')
```

### Marker edge color
* we can use the keyword argument markeredgecolor or in shorter form mec to set the color of the edges of the marker
* we can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers

```python
plt.plot(y_pts2,marker='o',ms=23,mec='red')
```

```python
plt.plot(y_pts2,marker='o',ms=23,mfc='red',mec='blue')
```

```python
# We can also use Hexadecimal values to give the color inside the marker
plt.plot(y_pts2, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

### Linestyle

* Use the keyword argument "linestyle" or shorter "ls", to change the plotted line

#### Shorter Syntax
    The line style can be written in a shorter syntax:

1. linestyle can be written as ls.

2. dotted can be written as :.

3. dashed can be written as --.

```python
plt.plot(y_pts2,ls='dotted')
```

```python
plt.plot(y_pts2,linestyle = 'dashed')
```

### Create labels for plot

* With pyplot we can able to create xlabel() and ylabel() functions to set a label for x and y axis

```python
x3 = np.array([10,23,44,66,54,9,87,89])
y3 = np.array([100,299,450,789,809,999,1043,1500])

plt.xlabel("Speed")
plt.ylabel("Distance")

plt.plot(x3,y3)
```

### Creating a title for plot -> title() function

```python
x3 = np.array([10,23,44,66,54,9,87,89])
y3 = np.array([100,299,450,789,809,999,1043,1500])

plt.title("Race")
plt.xlabel("Speed")
plt.ylabel("Distance")

plt.plot(x3,y3)
```

### Setting font properties for title and label

```python
font1 = {'family':'Arial','color':'darkred','size':15}
font2 = {'family':'times new roman','color':'Blue','size':20}

plt.title("Race in India",fontdict = font2)
plt.xlabel("Speed",fontdict = font1)
plt.ylabel("Distance",fontdict = font1)

plt.plot(x3,y3)
```

### Changing the position of the title -> loc

```python
plt.title("Race in India",loc = 'left',fontdict = font2)
plt.xlabel("Speed",fontdict = font1)
plt.ylabel("Distance",fontdict = font1)

plt.plot(x3,y3)
```

### Adding Grid lines to the plot

```python
plt.title("Race in India",loc = 'left',fontdict = font2)
plt.xlabel("Speed",fontdict = font1)
plt.ylabel("Distance",fontdict = font1)

plt.plot(x3,y3)
plt.grid()

plt.show()
```

### Specifying which grid lines need to be displayed it can be either x,y or both
### By default it'll be both

```python
plt.title("Race in India",loc = 'left',fontdict = font2)
plt.xlabel("Speed",fontdict = font1)
plt.ylabel("Distance",fontdict = font1)

plt.plot(x3,y3)
plt.grid(axis = 'x')

plt.show()
```

```python
plt.title("Race in India",loc = 'left',fontdict = font2)
plt.xlabel("Speed",fontdict = font1)
plt.ylabel("Distance",fontdict = font1)

plt.plot(x3,y3)
plt.grid(axis = 'y')

plt.show()
```

```python
plt.title("Race in India",loc = 'left',fontdict = font2)
plt.xlabel("Speed",fontdict = font1)
plt.ylabel("Distance",fontdict = font1)

plt.plot(x3,y3)
plt.grid(color = 'green',ls = 'dashed',lw = 0.5)

plt.show()
```

### Display Multiple plots

* Using subplot() function we can able to draw multiple plots in one figure

```python
# plot 1

x4 = np.array(list(range(1,5)))
y4 = np.array(list(range(6,10)))

# plot 2

x5 = np.array([2,1,5,3])
y5 = np.array([3,6,1,4])

plt.subplot(1,2,1) # Subplot takes 3 arguments row,col,index of the plot
plt.plot(x4,y4)

plt.subplot(1,2,2)
plt.plot(x5,y5)

# Now we've displayed 1 row and 2 column
```

```python
# If we need to have 2 rows and 1 column use the below syntax

plt.subplot(2,1,1)
plt.plot(x4,y4)

plt.subplot(2,1,2)
plt.plot(x5,y5)
```
