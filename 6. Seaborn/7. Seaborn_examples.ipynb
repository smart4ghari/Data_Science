```python
# import the required packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

### Scatter Plots

```python
# finding the correlation b/w salary and sales
plt.figure(figsize = (10,4),dpi = 200)
sns.scatterplot(x = 'salary',y = 'sales',data = df,hue = 'level of education',
               palette = 'Set2')
```

### Distribution Plots

```python
# Checking out the distribution of people salary

sns.rugplot(x = 'salary', data = df,height = 0.5)
```

```python
sns.displot(x = 'salary',data = df)
```

```python
# editing the bins

sns.set(style = 'dark')
sns.displot(x = 'salary', data = df, bins = 10,color = 'lightgreen',edgecolor = 'skyblue')
```

```python
sns.histplot(x = 'salary',data = df,bins = 10,kde = True)
```

```python
sns.kdeplot(x = 'salary',data = df)
```

```python
# Creating a sample ages
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
```

```python
sample_ages
```

```python
sns.displot(data = sample_ages,x = 'age',rug = True,kde = True)
```

```python
sns.kdeplot(data = sample_ages, x = 'age')
```

```python
sns.kdeplot(data = sample_ages, x = 'age',clip = [-10,90])
```

### Categorical plots

```python
df['division'].value_counts()
```

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.countplot(data = df,x = 'division')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize = (10,4),dpi = 500)
plt.set_cmap(cmap = 'tab10')
sns.countplot(data = df,x = 'level of education',hue = 'division')
```

```python
sns.set(style = 'white')
plt.figure(figsize = (10,4),dpi = 500)
sns.barplot(data = df, x = 'level of education',y = 'salary',estimator = np.mean,errorbar = 'sd',
           hue = 'division') 
plt.legend(bbox_to_anchor = (1.05,1))
# Here sd is std deviation
```

### Categorical plots

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.boxplot(data = df,y = 'reading score')
```

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.boxplot(data = df, y = 'reading score', x = 'parental level of education')
```

```python
# including with hue ( test preparation course )
plt.figure(figsize = (10,4), dpi = 200)

sns.boxplot(data = df, y = 'reading score', x = 'parental level of education', 
           hue = 'test preparation course')
```

```python
# Violin plot

plt.figure(figsize = (10,8),dpi = 200)

sns.violinplot(data = df, x = 'reading score', y = 'parental level of education')
```

```python
sns.swarmplot(data = df, x = 'math score',size = 2)
```

```python
plt.figure(figsize = (10,5),dpi = 200)
sns.swarmplot(data = df, x = 'math score', y = 'gender', size = 3)
```

```python
plt.figure(figsize = (10,5),dpi = 200)
sns.swarmplot(data = df, x = 'math score', y = 'gender', size = 3,
             hue = 'test preparation course',dodge = True)
```
